```python
import math
import numpy as np
import datetime
import pandas as pd
#import matplotlib.pyplot as plt

global_irr1=np.array([19.10827, 17.68327, 20.44018, 19.46753, 19.02541, 17.97876, 19.12089, 22.04202, 24.55404, 25.66702, 25.18559, 21.39024],dtype=np.float64)
tilt_angle=12 #Taken as User input in degrees
azimuth=0 #Taken as User input in degrees
interval=30 #User input in minutes and shows option for only 30 or 15 mins 
latitude=-10.7
global_irr=global_irr1[0]/3.6 #Value from Bom data

interval_length=24*(60/interval)
tilt_angle=-1*tilt_angle #For southern hemisphere
x=str(interval)+'T'

if(interval==30):
    time_series=pd.timedelta_range(0,periods=interval_length,freq="30T")
else:
     time_series=pd.timedelta_range(0,periods=interval_length,freq="15T")

days=np.array([91],dtype=np.int_) #Day number
#days=np.array([15,46,74,105,135,166,196,227,258,288,319,349],dtype=np.int_)
now=datetime.datetime.now()

output=pd.DataFrame({'Day': '15-Jan','Time interval':time_series})
#df = pd.DataFrame(data=[[91],[time_series]])
#df
#output2=pd.DataFrame({'Day': '15-Feb','Time interval':time_series})
#output3=pd.concat([output,output2],ignore_index=True)
#output3
hourofday = time_series / np.timedelta64(1, 'h') #gets hour of day from the time_series
hourofday=np.asarray(hourofday,dtype=np.float_)

#declination=np.full_like(hourofday,1)

declination_angle=23.45*np.sin(np.radians(360*(284+days[0])/365))
#declination_angle=np.full_like(hourofday,declination_angle)
sunset_angle=np.degrees(np.arccos(-1*np.arctan(np.radians(latitude))*np.arctan(np.radians(declination_angle))))
#sunset_angle=np.full_like(hourofday,sunset_angle)
#output['Declination Angle']=output.index
output['Declination Angle']=declination_angle
output['Sunset Angle']=sunset_angle
time_plot=['12 am','1 am','2 am','3 am','4 am','5 am','6 am', '7 am','8 am','9 am','10 am','11 am','12 pm','1 pm','2 pm','3 pm','4 pm','5 pm','6 pm', '7 pm', '8 pm', '9 pm','10 pm','11 pm']
hour_angle=np.full_like(hourofday,1)
#h0=np.full_like(hourofday,1)
h0=(24/math.pi)*3600*1368*(1+0.033*np.cos(2*math.pi*days[0]/365))*(np.cos(np.radians(latitude))*np.cos(np.radians(declination_angle))*np.sin(np.radians(sunset_angle))+np.radians(declination_angle)*np.sin(np.radians(latitude))*np.sin(np.radians(declination_angle)))*(0.000001/3.6)
kt=global_irr/h0
# hd is the Daily diffuse radiation and following are the conditions based on the value of kt
if(kt>0.17 and kt<0.75):
    hd=(1.188-2.272*kt+9.473*kt**2-21.865*kt**3+14.648*kt**4)*global_irr
elif (kt<0.17):
    hd=0.99*global_irr
elif (kt>0.75 and kt<0.8):
    hd=(0.632-0.5*kt)*global_irr
else:
    hd=0.2*global_irr
a=0.409+0.5016*np.sin(np.radians(sunset_angle)-(math.pi/3))
b=0.6609-0.4767*np.sin(np.radians(sunset_angle)-(math.pi/3))
rt=np.full_like(hourofday,1)
rd=np.full_like(hourofday,1)
rb=np.full_like(hourofday,1)
interval_rad=np.full_like(hourofday,1)
interval_diff=np.full_like(hourofday,1)
interval_beam=np.full_like(hourofday,1)
theta_z=np.full_like(hourofday,1)
theta_t=np.full_like(hourofday,1)
total_array=np.full_like(hourofday,1)
for i in range(len(hourofday)):
            
    if(abs(hourofday[i]-12)*15 > abs(sunset_angle)):
        hour_angle[i]=0
    else:
        hour_angle[i]=(hourofday[i]-12)*15
    rt[i]=math.pi/interval_length*(a+b*np.cos(np.radians(hour_angle[i])))*((np.cos(np.radians(hour_angle[i])))-np.cos(np.radians(sunset_angle)))/(np.sin(np.radians(sunset_angle))-np.radians(sunset_angle)*np.cos(np.radians(sunset_angle)))
    rd[i]=math.pi/interval_length*(np.cos(np.radians(hour_angle[i]))-np.cos(np.radians(sunset_angle)))/(np.sin(np.radians(sunset_angle))-np.radians(sunset_angle)*np.cos(np.radians(sunset_angle)))
    if(hour_angle[i]==0 and hourofday[i]!=12):
        interval_rad[i]=0
        interval_diff[i]=0
        theta_z[i]=0
        theta_t[i]=0
        rb[i]=0
        interval_beam[i]=interval_rad[i]-interval_diff[i]
        total_array[i]=0
    else:
        interval_rad[i]=global_irr*rt[i]*95/99
        interval_diff[i]=rd[i]*hd
        theta_z[i]=np.degrees(np.arccos(np.sin(np.radians(declination_angle))*np.sin(np.radians(latitude))+np.cos(np.radians(declination_angle))*np.cos(np.radians(latitude))*np.cos(np.radians(hour_angle[i]))))
        theta_t[i]=np.degrees(np.arccos(np.sin(np.radians(declination_angle))*np.sin(np.radians(latitude))*np.cos(np.radians(tilt_angle))-np.sin(np.radians(declination_angle))*np.cos(np.radians(latitude))*np.sin(np.radians(tilt_angle))*np.cos(np.radians(azimuth))+np.cos(np.radians(declination_angle))*np.cos(np.radians(latitude))*np.cos(np.radians(tilt_angle))*np.cos(np.radians(hour_angle[i]))+np.cos(np.radians(declination_angle))*np.sin(np.radians(latitude))*np.sin(np.radians(tilt_angle))*np.cos(np.radians(azimuth))*np.cos(np.radians(hour_angle[i]))+np.cos(np.radians(declination_angle))*np.sin(np.radians(tilt_angle))*np.sin(np.radians(azimuth))*np.sin(np.radians(hour_angle[i]))))
        rb[i]=np.cos(np.radians(theta_t[i]))/np.cos(np.radians(theta_z[i]))
        interval_beam[i]=interval_rad[i]-interval_diff[i]
        total_array[i]=interval_beam[i]*rb[i]+interval_diff[i]*((1+np.cos(np.radians(tilt_angle)))/2)
    

#Assigning to output
output['Hour Angle']=hour_angle
output['Daily extraterr']=h0
output['Daily diffuse radiation']=hd
output['Total radiation (kWh)']=interval_rad
output['Interval diffuse']=interval_diff
output['Interval beam']=interval_beam
#output['cos(Theta Z)']=theta_z
#output['cos(Theta T)']=theta_t
output['Total Array Radiation (kWh/m2/interval)']=total_array
output
#plt.plot(hourofday,total_array)
#a=time_series.to_pytimedelta()
#df = pd.DataFrame({"a": total_array}, index=time_series)
#df.plot()
#plt.xlabel('Time interval')
#plt.ylabel('Array irradition (kWh/m2/interval)')
#plt.title('Total Array radiation')
```
